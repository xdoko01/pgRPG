```python
TILE_RES = 64
ENT_VECT = (-1, 0)
ENT_POS = (340, 240)
VIEW_ANGLE = 60
NPC_POS = (50, 20)
VIEW_DIST = 300
```

```python
import math

VIEW_ANGLE = math.radians(VIEW_ANGLE//2)

dx = NPC_POS[0] - ENT_POS[0]
dy = NPC_POS[1] - ENT_POS[1]


dist = math.sqrt(dx*dx + dy*dy)
print(f'distance is: {dist}')

print(f'sin(0):{math.sin(math.radians(0))}, sin(30): {math.sin(math.radians(30))}, sinx: {dx/dist}')
print(f'NPC within the view: {"TRUE" if abs(dx/dist) > 0 and abs(dx/dist) < math.sin(math.radians(30)) else "FALSE" }')

view_dx = VIEW_DIST * math.sin(VIEW_ANGLE)
view_dy = VIEW_DIST * math.cos(VIEW_ANGLE)
```

```python
def is_in_view(source_pos_px: tuple, source_dir: tuple, target_pos_px: tuple, angle: int, max_dist: int) -> bool:

    angle = math.radians(angle // 2)

    dx = target_pos_px[0] - source_pos_px[0] 
    dy = target_pos_px[1] - source_pos_px[1]
    
    dist = math.sqrt(dx*dx + dy*dy)
    print(f'angle: {angle},\ndistance is: {dist},\ndx/dist: {dx/dist},\nmath.sin(angle): {math.sin(angle)},\ndx: {dx},\ndy: {dy},\nsource_dir: {source_dir}')
    
    if dist > max_dist: return False

    #(0,1)
    if abs(dx/dist) < math.sin(angle) and dy > 0 and source_dir[1] > 0:
        return True
    
    #(0,-1)
    if abs(dx/dist) < math.sin(angle) and dy < 0 and source_dir[1] < 0:
        return True
    
    #(1,0)
    if abs(dy/dist) < math.sin(angle) and dx > 0 and source_dir[0] > 0:
        return True

    #(-1,0)
    if abs(dy/dist) < math.sin(angle) and dx < 0 and source_dir[0] < 0:
        return True

    return False
```

```python
is_in_view(ENT_POS, ENT_VECT, NPC_POS, 60, 1000)
```

```python
from ipycanvas import MultiCanvas

multi_canvas = MultiCanvas(4, size=(640, 256))

# Basic grid
multi_canvas[0].fill_style = 'grey'
multi_canvas[0].stroke_style = 'black'
for y in range(5):
    for x in range(10):
        multi_canvas[0].stroke_rect(x*TILE_RES, y*TILE_RES, TILE_RES)

# Entity line
multi_canvas[1].fill_style = 'grey'
multi_canvas[1].stroke_style = 'black'

multi_canvas[1].begin_path()
multi_canvas[1].move_to(ENT_POS[0], ENT_POS[1])
multi_canvas[1].line_to(ENT_POS[0] + ENT_VECT[0], ENT_POS[1] + ENT_VECT[1])
multi_canvas[1].stroke()

# NPC line
multi_canvas[2].fill_style = 'blue'
multi_canvas[2].stroke_style = 'black'

multi_canvas[2].begin_path()
multi_canvas[2].move_to(ENT_POS[0], ENT_POS[1])
multi_canvas[2].line_to(NPC_POS[0], NPC_POS[1])
multi_canvas[2].stroke()

# The View

multi_canvas[3].fill_style = 'lime'
multi_canvas[3].stroke_style = 'green'

multi_canvas[3].begin_path()
multi_canvas[3].move_to(ENT_POS[0], ENT_POS[1])

if ENT_VECT[1] > 0:
    multi_canvas[3].line_to(ENT_POS[0] + view_dx, ENT_POS[1] + view_dy)
    multi_canvas[3].move_to(ENT_POS[0], ENT_POS[1])
    multi_canvas[3].line_to(ENT_POS[0] - view_dx, ENT_POS[1] + view_dy)
elif ENT_VECT[1] < 0:
    multi_canvas[3].line_to(ENT_POS[0] + view_dx, ENT_POS[1] - view_dy)
    multi_canvas[3].move_to(ENT_POS[0], ENT_POS[1])
    multi_canvas[3].line_to(ENT_POS[0] - view_dx, ENT_POS[1] - view_dy)
elif ENT_VECT[0] > 0:
    multi_canvas[3].line_to(ENT_POS[0] + view_dx, ENT_POS[1] - view_dy)
    multi_canvas[3].move_to(ENT_POS[0], ENT_POS[1])
    multi_canvas[3].line_to(ENT_POS[0] + view_dx, ENT_POS[1] + view_dy)
elif ENT_VECT[0] < 0:
    multi_canvas[3].line_to(ENT_POS[0] - view_dy, ENT_POS[1] - view_dx)
    multi_canvas[3].move_to(ENT_POS[0], ENT_POS[1])
    multi_canvas[3].line_to(ENT_POS[0] - view_dy, ENT_POS[1] + view_dx)

multi_canvas[3].stroke()

multi_canvas
```
